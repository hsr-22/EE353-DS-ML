We summarize our data across various columns based on their characteristics and their potential relationship to our prediction variable. To achieve this, we utilize different SQL queries. The MAX query retrieves the most frequently occurring value in a column for each ZIP code. The SUM query calculates the total of all values in a column for a specific ZIP code. The AVG query computes the average of all values in a column for a given ZIP code.

```python
query = """ SELECT ZIP, MAX("DEVELOPMENT STAGE"), MAX("CONSTRUCTION TYPE"),SUM("PROJECT TOTAL UNITS"),MAX("HOUSING TYPE"),MAX("SUPPORTIVE HOUSING"), AVG("SH UNITS PER SITE"),AVG("LAHD FUNDED"),AVG("TAX EXEMPT CONDUIT BOND"),AVG(LEVERAGE),AVG("TDC"),SUM("JOBS")
FROM Housing
GROUP BY ZIP;""" # Query to select and summarise particular columns in particular ways. 

summ1=pd.read_sql_query(query, conn) # Dataframe to store the outcome of the query executed
summ1
```

```python
query = """ SELECT ZIP, AVG("score")
FROM Health
GROUP BY ZIP;""" # Query run on the Health table to get the average score for each ZIP code

summ2=pd.read_sql_query(query, conn) # Dataframe to store the outcome of the query executed
summ2
```

Now we create new tables to store the summary data, which will actually be used for finding correlations and analysis.

```python
# Creating summary tables in the database so that they can be combined later

cursor3=conn.cursor()
query1=""" CREATE TABLE IF NOT EXISTS Health_Summary
(ZIP INTEGER, "Average Score" FLOAT
    )"""
    
cursor3.execute(query1)

query2="""CREATE TABLE IF NOT EXISTS Housing_Summary
(ZIP INTEGER,
"Maximum Development Type"            INTEGER,
"Maximum Construction Type"            INTEGER,
"Total Project Units"          INTEGER,
"Maximum Housing Type"                INTEGER,
"Maximum Houses Supported"           INTEGER,
"Average SH units per site"          FLOAT,
"Average LAHD funded"               FLOAT,
"Average Tax exempted"    FLOAT,
"Average Leverage"                     FLOAT,
"Average TDC"                        FLOAT,
"Total Jobs"                       FLOAT)
"""
cursor3.execute(query2)

summ1.to_sql('Housing_Summary', conn, if_exists="replace", index=False)
summ2.to_sql('Health_Summary', conn, if_exists="replace", index=False)
```

Combining both of our summary tables at all those values of the ZIP code which exist in both of them. 

```python
# Selecting all the columns of both the summary tables and combining both of them for those values where the ZIP code in both the tables match

query = """ SELECT *                
FROM Housing_Summary AS hp
JOIN Health_Summary AS hi ON hp.ZIP = hi.ZIP;"""

result = pd.read_sql_query(query, conn)
result # Dataframe stores our final table for analysis
```

```python
result=result.sort_values('AVG("score")') # Sorting the dataframe based on the health score values
result
```

We test the following hypothesis now:

#### Alternative Hypothesis ($H_1$): 
The statistics of the affordable housing projects is related to the health inspection scores of the restaurants for a particular ZIP code

#### Null Hypothesis ($H_0$):
There is no relation between the health inspection scores and the housing data at any ZIP code

Since our data is non-parametric (we haven't assumed any distributions to be fitting) and all of it is completely numeric without any category, we use the Spearman Correlation Test for testing our hypothesis. We set the p-value threshold at 0.05 to check for the best correlation.

The block below creates a correlation heatmap between all possible combinations of variables using the given test, and it helps us to easily observe the correlation factor values. The more the value of the correlation factor, the stronger the correlation between the variables. We need to check the correlation of the health score variable with other variables present in the housing data summary table. 

```python
plt.figure(figsize=(12,12), dpi= 80) # specifying figure details
sns.heatmap(result.corr('spearman'), xticklabels=result.corr('spearman').columns, yticklabels=result.corr('spearman').columns, cmap='RdYlGn', center=0, annot=True) # creating heatmap having correlation values for column pairs, using the spearman correlation test

# Decorations
plt.title('Correlogram of Housing and Health Scores Report', fontsize=22) # specifying title for plot
plt.xticks(fontsize=12) # specifying font size for ticks at x-axis
plt.yticks(fontsize=12) # specifying font size for ticks at y-axis
plt.tight_layout()
plt.show()
```

From the above chart, on observing the row (or column) for the average score, we can see it has the strongest positive correlation with the Average Tax Exempt Conduit Bond and then the Maximum type of construction done in a given ZIP code. It also has a strong negative correlation with the Average Leverage amount given to a particular ZIP code.

```python
from scipy.stats import spearmanr

cor,p=spearmanr(result['AVG("TAX EXEMPT CONDUIT BOND")'], result['AVG("score")'])
print("Correlation coefficient = ",cor)
print("p-value = ", p)

cor1,p1=spearmanr(result['MAX("CONSTRUCTION TYPE")'], result['AVG("score")'])
print("Correlation coefficient = ",cor1)
print("p-value = ", p1)

# Negative Correlation
cor2,p2=spearmanr(result['AVG(LEVERAGE)'], result['AVG("score")'])
print("Correlation coefficient = ",cor2)
print("p-value = ", p2)

# Random variable with bad correlation taken for comparison
cor3,p3=spearmanr(result['MAX("HOUSING TYPE")'], result['AVG("score")'])
print("Correlation coefficient = ",cor3)
print("p-value = ", p3)
#[12]
```

The correlation factor for the random uncorrelated variable is significantly lower compared to the other related variables, while the p-values show the opposite trend. Therefore, we cannot reject the null hypothesis for the last variable. However, for the other three variables, we can reject the null hypothesis and accept the alternative hypothesis.

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from scipy.stats import ttest_ind

# Load the CSV file
file_path = 'datafile.csv'  # Update with the correct file path
data = pd.read_csv(file_path)

# Convert 'Year' and 'Month' columns to a datetime format
data['Month'] = data['Month'].str.strip()
data['Date'] = pd.to_datetime(data['Year'].astype(str) + data['Month'], format='%Y%B', errors='coerce')

# Check for any issues in date conversion
if data['Date'].isnull().any():
    print("Warning: Some dates could not be parsed. Check the 'Month' column for inconsistencies.")
    print(data[data['Date'].isnull()][['Year', 'Month']])
```

```python
# Step 1: Exploratory Data Analysis (EDA)
# 1.1 Check for missing values
missing_values = data.isnull().sum()
print("Missing Values:\n", missing_values)

# 1.2 Descriptive statistics
descriptive_stats = data.describe()
print("Descriptive Statistics:\n", descriptive_stats)

# 1.3 Correlation matrix
# Exclude non-numeric columns
numeric_data = data.select_dtypes(include=[float, int])
correlation_matrix = numeric_data.corr()
```

```python
# Step 2: Data Visualization
# 2.1 Correlation Heatmap
plt.figure(figsize=(14, 12))
sns.heatmap(correlation_matrix, annot=False, linewidths=0.5, square=True)
plt.title('Correlation Heatmap of Index Categories')
plt.show()

# 2.2 Time Series Plot for 'Vegetables' and 'Cereals and products'
plt.figure(figsize=(14, 8))
plt.plot(data['Date'], data['Vegetables'], label='Vegetables Index', color='green')
plt.plot(data['Date'], data['Cereals and products'], label='Cereals and Products Index', color='orange')
plt.xlabel('Date')
plt.ylabel('Index Value')
plt.title('Trend of Vegetables and Cereals over Time')
plt.legend()
plt.grid(True)
plt.show()
```

```python
# Step 3: Hypothesis Testing
# Hypothesis: The index for 'Vegetables' has a significantly different trend compared to 'Cereals and products' over the years.

# Group data by year to compare yearly averages
veg_yearly_avg = data.groupby('Year')['Vegetables'].mean()
cereals_yearly_avg = data.groupby('Year')['Cereals and products'].mean()

# Perform a t-test to compare the means of the two groups
t_stat, p_value = ttest_ind(veg_yearly_avg, cereals_yearly_avg, equal_var=False)

print(f"{p_value}")

# Interpretation
if p_value < 0.05:
    print("There is a statistically significant difference between the trends of Vegetables and Cereals indices.")
else:
    print("There is no statistically significant difference between the trends of Vegetables and Cereals indices.")
```
